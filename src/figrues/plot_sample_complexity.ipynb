```python
import math
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc

rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
mpl.rcParams['savefig.dpi'] = 1200
mpl.rcParams['text.usetex'] = True  # not really needed

import matplotlib.ticker as mticker

training_subsets = [0.2, 0.4, 0.6, 0.8]
linear_error = [2.11, 8, 3.45, 1.42]
kernel_error = [1.45, 1.51, 1.39, 1.17]

fig, ax = plt.subplots(figsize=(6, 5))

plt.scatter(training_subsets, linear_error, color='blue', alpha=0.7, s=400)
plt.scatter(training_subsets, kernel_error, color='red', alpha=0.7, s=400)
plt.plot(training_subsets, linear_error, color='blue', linewidth=4, ls='--', label=f"Linear")
plt.plot(training_subsets, kernel_error, color='red', linewidth=4, ls='--', label=f"Kernel")

#plt.xscale('log')  # Hessians vary widely, log scale makes it clearer
plt.xlabel(r"$\mathrm{Subset~split}$", fontsize=40)
plt.xticks([0.2, 0.4, 0.6, 0.8], fontsize=40)
plt.ylabel(r"$\mathrm{Error}$", fontsize=40)
plt.yticks([1, 4, 7,10], fontsize=40)
#plt.title("Connection between Train Hessian and Datamodels Normalized Error", fontsize=40)
plt.grid(lw=0.4, ls=":")
plt.tight_layout()

plt.savefig("sample_complexity_modular.pdf", format="pdf", dpi=1200)
plt.show()
```
