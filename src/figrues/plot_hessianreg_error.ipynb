```python
import math
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc

rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
mpl.rcParams['savefig.dpi'] = 1200
mpl.rcParams['text.usetex'] = True  # not really needed

import matplotlib.ticker as mticker

training_epochs = [500, 1000, 1500, 2000]
sgd_error = [0.75, 1.42, 2.86, 5.32]
sam_error = [2.17, 2.17, 2.16, 2.20]
nso_error = [0.91, 1.32, 1.82, 1.89]

fig, ax = plt.subplots(figsize=(6, 5))

plt.scatter(training_epochs, sgd_error, color='blue', alpha=0.7, s=400)
plt.scatter(training_epochs, sam_error, color='red', alpha=0.7, s=400)
plt.scatter(training_epochs, nso_error, color='green', alpha=0.7, s=400)
plt.plot(training_epochs, sgd_error, color='blue', linewidth=4, ls='--', label=r"$\mathrm{SGD}$")
plt.plot(training_epochs, sam_error, color='red', linewidth=4, ls='--', label=r"$\mathrm{SAM}$")
plt.plot(training_epochs, nso_error, color='green', linewidth=4, ls='--', label=r"$\mathrm{NSO}$")

plt.legend(fontsize=30, loc='upper right', bbox_to_anchor=(0.56, 0.98), borderaxespad=0.)

plt.xlabel(r"$\mathrm{Epochs}$", fontsize=40)
plt.xticks([500, 1000, 1500, 2000], fontsize=40)
ax.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
ax.xaxis.offsetText.set_fontsize(40)
plt.ylabel(r"$\mathrm{Error}$", fontsize=40)
plt.yticks([2, 4, 6], fontsize=40)
#plt.title("Connection between Train Hessian and Datamodels Normalized Error", fontsize=40)
plt.grid(lw=0.4, ls=":")
plt.tight_layout()

plt.savefig("hessianreg_error.pdf", format="pdf", dpi=1200)
plt.show()
```
