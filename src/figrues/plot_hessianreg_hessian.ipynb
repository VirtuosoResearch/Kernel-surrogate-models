```python
import math
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc

rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
mpl.rcParams['savefig.dpi'] = 1200
mpl.rcParams['text.usetex'] = True  # not really needed

import matplotlib.ticker as mticker

training_epochs = [500, 1000, 1500, 2000]
sgd_hessian = [2e5, 1e6, 5e6, 1e7]
sam_hessian = [8e4, 1e5, 9e4, 1e4]
nso_hessian = [2e4, 1e4, 1e3, 1e3]

fig, ax = plt.subplots(figsize=(6, 5))

plt.scatter(training_epochs, sgd_hessian, color='blue', alpha=0.7, s=400)
plt.scatter(training_epochs, sam_hessian, color='red', alpha=0.7, s=400)
plt.scatter(training_epochs, nso_hessian, color='green', alpha=0.7, s=400)
plt.plot(training_epochs, sgd_hessian, color='blue', linewidth=4, ls='--', label=f"SGD")
plt.plot(training_epochs, sam_hessian, color='red', linewidth=4, ls='--', label=f"SAM")
plt.plot(training_epochs, nso_hessian, color='green', linewidth=4, ls='--', label=f"NSO")

#plt.xscale('log')  # Hessians vary widely, log scale makes it clearer
plt.xlabel(r"$\mathrm{Epochs}$", fontsize=40)
plt.xticks([500, 1000, 1500, 2000], fontsize=40)
plt.ylabel(r"$\mathrm{Hessian}$", fontsize=40)
plt.yscale('log')
plt.yticks([1e3,1e5,1e7], fontsize=40)
#plt.title("Connection between Train Hessian and Datamodels Normalized Error", fontsize=40)
plt.grid(lw=0.4, ls=":")
plt.tight_layout()

plt.savefig("hessianreg_hessian.pdf", format="pdf", dpi=1200)
plt.show()
```
